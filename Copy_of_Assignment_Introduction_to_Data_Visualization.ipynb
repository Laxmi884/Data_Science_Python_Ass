# <b><u>Problem 2 </u></b>

### Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```
# Your code here
ax = df.groupby(['month_number'])['facecream','facewash'].mean().plot.bar(figsize=(12,6),width=0.8)
plt.xlabel('Month'),
plt.ylabel("Units_Sold")
```

# <b><u>Problem 3</u></b>

### Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

```
# Your code here
ap = df.loc[:,['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer','month_number']]
ap = ap.set_index('month_number')
plt.plot(ap['facecream'],'ro-', linewidth=2, markersize=5)
plt.plot(ap['facewash'],'go--', linewidth=2, markersize=5)
plt.plot(ap['toothpaste'],'yo-', linewidth=2, markersize=5)
plt.plot(ap['bathingsoap'],'bo-', linewidth=2, markersize=5)
plt.plot(ap['shampoo'],'co-', linewidth=2, markersize=5)
plt.plot(ap['moisturizer'],'ko-', linewidth=2, markersize=5)

plt.legend(['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'])

plt.title("Monthly Unit Sales For Each Product",fontsize=15)
plt.xlabel("Month Number",fontsize=15)
plt.ylabel("Unit Sales",fontsize=15)

plt.rcParams['figure.figsize']=(10,15)
```
