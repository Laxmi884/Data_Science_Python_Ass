# <b><u> Problem 1: </u> Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor. </b>

# <b><u> Problem 2 :</u> Top 3 directors prefer which actors the most?

```python
# Your code here 
def find(li):
  it = 0
  if type(li)!=float:
    li = ast.literal_eval(li)
    while True:
      if li[it]['job']!='Director':
        it+=1
      else:
        return li[it]['name']
      



df['Director_name'] =df.apply(lambda x: find(x['crew']),axis=1)
gp_dit = df.groupby('Director_name')['popularity'].mean().sort_values(ascending=False).reset_index()
gp_dit.head(3)
```

```python
#Actors prefered by tom miller
pop_dic_act1 = list(df[df['Director_name']==gp_dit['Director_name'][0]]['new_cast'])
pop_dic_act1
```

```python
#actors preferred by james gun
act2 = list(df[df['Director_name']==gp_dit['Director_name'][1]]['new_cast'])
act2
```

```python
df_na = df[~df['crew'].isna()]
df_na.head()
```

```python
from traitlets.traitlets import literal_eval
def get_producers_and_director_list(string_of_crew_list):
  try:
    producer_list=[]
    director_list=[]
    crew_list = literal_eval(string_of_crew_list)
    for elm in crew_list:
      if elm['job']=='Producer':
        producer_list.append(elm['name'])
      if elm['job']=='Director':
        director_list.append(elm['name'])
    return producer_list,director_list
  except:
    return [],[]
```

```python
df[['producers','directors']]=df.apply(lambda s: get_producers_and_director_list(s['crew']), axis=1, result_type="expand")
df.head()
```
