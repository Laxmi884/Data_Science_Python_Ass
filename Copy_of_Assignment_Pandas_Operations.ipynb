# <b> Problem 8 : Create a new column for the updated price. This updated price will be calculated on top of the given price - if the engine is in front, price will be same else if the engine is in rear, price will be doubled </b>


```python
# pd.set_option('max_rows', 99999)
# pd.set_option('max_colwidth', 99999)
# pd.describe_option('max_colwidth')
pd.set_option('display.max_rows', None)
# Your code here. Use apply and lambda
def cal_eng_price(engine,price):
  if engine=='front':
    return 1*price
  else:
    return 2*int(price)

automobile_df["updated_price"]=automobile_df.apply(lambda x : cal_eng_price(x['engine-location'],x['price']),axis=1)
automobile_df
```

# <b> Problem 10 : Create a new column which stores the number of doors in a car as integers. 

```python
# Your code here. Use apply
automobile_df['num-of-doors'].value_counts()
def con_dorr(door_no):
  if door_no == 'four':
    door_no = 4
    return door_no
  elif door_no =='two':
    door_no =2
    return door_no

automobile_df['No_of_doors'] = automobile_df.apply(lambda x:con_dorr(x['num-of-doors']),axis=1)
automobile_df
```

# <b>Problem 11 : Concatenate the two data frames given below firstly row wise and secondly column wise.</b>
